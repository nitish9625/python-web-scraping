```python
import requests
from bs4 import BeautifulSoup
```

```python
names = []
phones = []
emails = []
websites = []

page_num = input("Enter a number: ")
for i in range(1, int(page_num)+1):
    url = "https://www.dica.gov.mm/en/myindy/private-enterprises?page="+str(i)
    req = requests.get(url)
    content = BeautifulSoup(req.content, 'html.parser')
#     print(content)
    name = content.find_all('div',{'class':'company-name'})
    phone = content.find_all('div', {'class':"company-phone"})
    email = content.find_all('div', {'class':'company-email'})
    website = content.find_all('div', {'class', 'company-website'})
    print('Name in company'+str(i))
    print(len(name))
    
    for i in name:
        names.append(i.text)
    for i in phone:
        phones.append(i.text)
    for i in email:
        emails.append(i.text)
    for i in website:
        websites.append(i.text)
```

```python
for i in names:
    print(i)
```

```python
import pandas as pd
```

```python
data ={'company name':names,'company phone':phones, 'company email':emails, 'company website':websites}
df = pd.DataFrame(data)
df
```

```python
df.to_csv('webscraping.csv')
```
